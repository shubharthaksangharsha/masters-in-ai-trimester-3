### 2. Create a histogram of the average IMDB rating (IMDB Rating) for all TV series in the dataset. Hint: Remember to add axis labels. Note that we are asking for the series’ rating, not how many votes it received.

```python
#import the library 
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['IMDB_Rating'], bins=10, color='green', alpha=0.7)
plt.xlabel('IMDB Rating')
plt.ylabel('Number of TV Series-Frequency')
plt.title('Histogram of IMDB Ratings for TV Series')
plt.tight_layout()
plt.show();
```

### 3. Calculate the Mean IMDB Rating for All TV Series

```python
imdb_mean_rating = df['IMDB_Rating'].mean() #calculating the mean using mean function 

print(f"The mean IMDB rating for all TV series is: {round(imdb_mean_rating, 3)}") #printing the mean of IMDB rating till 3 decimals-up
```

### 4. Finding TV Series with most Votes (IMDB rating above 8)

```python
high_rating_series = df[df['IMDB_Rating'] > 8] #getting the tv series which have imdb rating higher than 8 
high_rating_series
```

### The above new `high_rating_series` dataframe have IMDB rating more than 8

#### Now, we will find which have highest votes among them using `idxmax` method. 

```python
most_voted_series = high_rating_series.loc[high_rating_series['No_of_Votes'].idxmax()]
most_voted_series
```

```python
# Find the series with the most votes
most_voted_series = high_rating_series.loc[high_rating_series['No_of_Votes'].idxmax()] #idxmax (get us the location of maximum votes)

# Print the result
print(f"The series with the most votes among those with an IMDB rating above 8 is: {most_voted_series['Series_Title']}")
```

### 5. Now find the “Crime” series with the most votes among series with an IMDB rating above 8

```python
crime_series = df[(df['IMDB_Rating'] > 8) & (df['Genre'].str.contains('Crime', case=False))] #getting the crime series which have IMDB rating more than 8 
most_voted_crime_series = crime_series.loc[crime_series['No_of_Votes'].idxmax()] #finding the series with most votes among crime series
print(f"The 'Crime' series with the most votes among those with an IMDB rating above 8 is: {most_voted_crime_series['Series_Title']}") #printing the series
```
